---In our comprehensive analysis, we examined data from 15 schools and over 32,000 students. Our goal was to determine which schools, categorized by type (District or Charter), exhibited the highest performance in math and reading scores.

Interestingly, we found that a higher budget does not necessarily equate to higher test scores. This challenges the common assumption that increased funding directly improves academic performance.

Furthermore, our analysis revealed a notable trend: Charter schools consistently outperformed District schools in terms of test scores, despite the latter often having larger budgets. This suggests that factors beyond financial resources, such as teaching methods, school culture, or curriculum design, may play a significant role in student achievement.

In conclusion, our study underscores the complexity of educational outcomes and the importance of considering multiple factors when evaluating school performance. It also prompts further investigation into the practices employed by Charter schools that contribute to their studentsâ€™ high test scores.

```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load (Remember to Change These)
school_data_to_load = Path("complete schools.csv")
student_data_to_load = Path("complete_students.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

```python
# Calculate the total number of unique schools
school_count = school_data['school_name'].nunique()
school_count
```

```python
# Calculate the total number of students
student_count = student_data['student_name']. nunique()
student_count
```

```python
# Calculate the total budget
total_budget = school_data_complete['budget'].sum()
total_budget
```

```python
# Calculate the average (mean) math score
average_math_score = school_data_complete['math_score'].mean()
average_math_score
```

```python
# Calculate the average (mean) reading score
average_reading_score = school_data_complete['reading_score'].mean()
average_reading_score
```

```python
# Use the following to calculate the percentage of students who passed math (math scores greather than or equal to 70)
passing_math_count = school_data_complete[(school_data_complete["math_score"] >= 70)].count()["student_name"]
passing_math_percentage = passing_math_count / float(student_count) * 100
passing_math_percentage
```

```python
# Calculate the percentage of students who passed reading (hint: look at how the math percentage was calculated)  
passing_reading_count = school_data_complete[(school_data_complete["reading_score"] >=70)].count()["student_name"]
passing_reading_percentage = passing_reading_count / (student_count) * 100
passing_reading_percentage
```

```python
# Use the following to calculate the percentage of students that passed math and reading
passing_math_reading_count = school_data_complete[
    (school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)
].count()["student_name"]
overall_passing_rate = passing_math_reading_count /  float(student_count) * 100
overall_passing_rate
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame
district_summary = pd.DataFrame({
    'Total Schools': [school_count],
    'Total Students': [student_count],
    'Total Budget': [total_budget],
    'Average Math Score': [average_math_score],
    'Average Reading Score': [average_reading_score],
    '% Passing Math': [passing_math_percentage],
    '% Passing Reading': [passing_reading_percentage],
    '% Overall Passing': [overall_passing_rate]})

# Formatting
district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)

# Display the DataFrame
district_summary
```

## School Summary

```python
# Use the code provided to select all of the school types
school_types = school_data_complete.groupby('school_name')['type'].first()

school_types_df = school_types.to_frame()

school_types_df.columns = ['School Type']

school_types_df
```

```python
# Calculate the total student count per school
per_school_counts = school_data_complete.groupby('school_name')['student_name'].count()
per_school_counts
per_school_counts_df = pd.DataFrame(per_school_counts)
per_school_counts_df.columns = ['Total Students']
per_school_counts_df
```

```python
# Calculate the total school budget and per capita spending per school
per_school_budget = school_data_complete.groupby('school_name')['budget'].sum()
per_school_capita = per_school_budget / per_school_counts
school_summary_spending = pd.DataFrame({
    'Total Budget': per_school_budget,
    'Per Capita Spending': per_school_capita
})
school_summary_spending
```

```python
# Calculate the average test scores per school
per_school_math = school_data_complete.groupby('school_name')['math_score'].mean()
per_school_reading = school_data_complete.groupby('school_name')['reading_score'].mean()

average_test_per_school = pd.DataFrame({
    'Average Math Score Per School': per_school_math,
    'Average Reading Score Per School': per_school_reading
})
average_test_per_school

```

```python
# Calculate the number of students per school with math scores of 70 or higher
students_passing_math = school_data_complete[school_data_complete['math_score'] >= 70].groupby('school_name')['student_name'].count()

students_passing_math_df = pd.DataFrame(students_passing_math)
students_passing_math_df.columns = ['Students Passing Math']

students_passing_math_df
```

```python
students_passing_reading = school_data_complete[school_data_complete['reading_score'] >= 70].groupby('school_name')['student_name'].count()

students_passing_reading_df = pd.DataFrame(students_passing_reading)
students_passing_reading_df.columns = ['Students Passing Reading']

students_passing_reading_df
```

```python
# Use the provided code to calculate the number of students per school that passed both math and reading with scores of 70 or higher
students_passing_math_and_reading = school_data_complete[
    (school_data_complete["reading_score"] >= 70) & (school_data_complete["math_score"] >= 70)
]
school_students_passing_math_and_reading = students_passing_math_and_reading.groupby(["school_name"]).size()
school_students_passing_math_and_reading
df_students_passing_math_and_reading = pd.DataFrame(school_students_passing_math_and_reading, columns=['Number of Students Passing Both'])
df_students_passing_math_and_reading
```

```python
passing_math_rate_per_school = (school_students_passing_math / per_school_counts) * 100

passing_reading_rate_per_school = (students_passing_reading / per_school_counts) * 100

students_passing_both = school_data_complete[
    (school_data_complete["reading_score"] >= 70) & (school_data_complete["math_score"] >= 70)
]

school_students_passing_both = students_passing_both.groupby(["school_name"])['student_name'].count()

passing_both_rate_per_school = (school_students_passing_both / per_school_counts) * 100

students_success_rates_per_school = pd.DataFrame({
    'students passing math per school': passing_math_rate_per_school,
    'students passing reading per school': passing_reading_rate_per_school,
    'students passing math and reading percent': passing_both_rate_per_school
})

students_success_rates_per_school
```

```python
# Create a DataFrame called `per_school_summary` with columns for the calculations above.
per_school_summary = pd.DataFrame({
    'School Type': school_types_df['School Type'],
    'Total Students': per_school_counts_df['Total Students'],
    'Total School Budget': school_summary_spending['Total Budget'],
    'Per Student Budget': school_summary_spending['Per Capita Spending'],
    'Average Math Score': average_test_per_school['Average Math Score Per School'],
    'Average Reading Score': average_test_per_school['Average Reading Score Per School'],
    '% Passing Math': students_passing_math_and_reading_per_school['students passing math per school'],
    '% Passing Reading': students_passing_math_and_reading_per_school['students passing reading per school'],
    '% Overall Passing': students_success_rates_per_school['students passing math and reading percent']
})

per_school_summary['Total School Budget'] = per_school_summary['Total School Budget'].map("${:,.2f}".format)
per_school_summary['Per Student Budget'] = per_school_summary['Per Student Budget'].map("${:,.2f}".format)

per_school_summary
```
